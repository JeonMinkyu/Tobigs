# 1. Load Data

```python
import nltk
# NLTK와 KoNLPy에서 제공하는 문서를 쓸것임
# 영어: Jane Austen의 소설 Emma
# 한국어: 대한민국 국회 제 1809890호 의안
```

```python
#English
from nltk.corpus import gutenberg   # Docs from project gutenberg.org
files_en = gutenberg.fileids()      # Get file ids
doc_en = gutenberg.open('austen-emma.txt').read()
```

```python
#Korean
from konlpy.corpus import kobill    # Docs from pokr.kr/bill
files_ko = kobill.fileids()         # Get file ids
doc_ko = kobill.open('1809890.txt').read()
```

# 2. Tokenize

* 문서를 토큰으로 나누는 방법은 다양하다.
* 여기서는 영어에는 nltk.regexp_tokenize, 한국어에는 konlpy.tag.Twitter.morph를 사용해보자.

```python
#E
from nltk import regexp_tokenize
pattern = r'''(?x) ([A-Z]\.)+ | \w+(-\w+)* | \$?\d+(\.\d+)?%? | \.\.\. | [][.,;"'?():-_`]'''
tokens_en = regexp_tokenize(doc_en, pattern)
tokens_en
```

```python
# K
from konlpy.tag import Twitter
t = Twitter()
tokens_ko = t.morphs(doc_ko)
tokens_ko
```

# 3. Load tokens with nltk.Text()

```python
#E
en = nltk.Text(tokens_en)
en
```

```python
#K
# ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
ko = nltk.Text(tokens_ko)
ko
```

## 1) Tokens

```python
# E
print(len(en.tokens))       # rnumber of tokens (document length)
print(len(set(en.tokens)))  # number of unique tokens
en.vocab()                  # frequency distribution
```

```python
# K
print(len(ko.tokens))       # number of tokens (document length)
print(len(set(ko.tokens)))  # rnumber of unique tokens
ko.vocab()                  # frequency distribution
```

## 2) Plot frequency distributions

```python
#E
en.plot(50)     # Plot sorted frequency of top 50 tokens
```

```python
#K
ko.plot(50)     # Plot sorted frequency of top 50 tokens
```
