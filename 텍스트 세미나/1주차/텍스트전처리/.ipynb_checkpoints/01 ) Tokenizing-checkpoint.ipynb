# 1. Read document

```python
#Korean
from konlpy.corpus import kobill    # Docs from pokr.kr/bill
files_ko = kobill.fileids()         # Get file ids
doc_ko = kobill.open('1809890.txt').read()  # 대한민국 국회 의안 제 1809890호
doc_ko[:1000]
```

# 2. Tokenize

* 문서를 토큰으로 나누는 방법은 다양하다.
* 한국어 토큰 나누기는 Twitter.morph를 사용해보자.

```python
# K
from konlpy.tag import Twitter
t = Twitter()
tokens_ko = t.morphs(doc_ko)
tokens_ko
```

# 3. Load tokens with nltk.Text()
* nltk.Text()를 통해 문서 하나를 편리하게 탐색 할 수 있다

```python
# K
# ko = nltk.Text(tokens_ko, name='대한민국 국회 의안 제 1809890호')
import nltk
ko = nltk.Text(tokens_ko)

ko
```

## 1) Tokens

```python
# K
print(len(ko.tokens))       # number of tokens (document length)
print(len(set(ko.tokens)))  # rnumber of unique tokens
ko.vocab()                  # frequency distribution
```

## 2) Plot frequency distributions

```python
# K
ko.plot(50)     # Plot sorted frequency of top 50 tokens
```

## 3) count

```python
#k
print(ko.count('육아휴직'))
print(ko.count('초등학교'))
print(ko.count('공무원'))
```

## 4) Dispersion plot
* 텍스트 내 단어 사용 빈도와 위치를 분산 그래프로 보여준다
* plot 쓰려면 NumPy 설치되어있어야하는데 투빅이는 당연히 깔려있겠ㅈ지?!

```python
#K
ko.dispersion_plot(['육아휴직', '초등학교', '공무원'])
```

## 5) Concordance (색인)
* 단어가 쓰인 모든 경우의 문맥을 찾아준다
* 책의 색인을 보면 책에 나온 단어들이 나열되어있고, 쓰여있는 페이지가 나와있다
* nltk의 concordance는 텍스트에서 주어진 단어가 쓰인 문맥을 다 찾아준다

```python
ko.concordance('초등학교')
```

## 6) Similar
* 주어진 단어와 비슷한 환경에서 쓰인 단어를 찾아줌
* 뜻이 비슷한 것 xxxxxx, 순전히 쓰이는 환경이 비슷한 단어를 찾아주는 거임.....

```python
ko.similar('자녀')
```

## 7) Collocations (연어)
* 같이 쓰인 단어를 찾을 수 있다

```python
ko.collocations()
```
